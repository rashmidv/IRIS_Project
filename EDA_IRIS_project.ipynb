```python
def which_iris():
    PL= float(input("Enter petal_length"))
    PW= float(input("Enter petal width"))
    try:
        if(PL.dtype == str or PW.dtype == str)
        
    except:
        print("Please enter integer value greater 0")
        
    if(0<PL<2 and 0<PW <0.7):
        print("The flower is of setosa")
    elif(3<PL<5.2 and 1<PW<1.7):
        print("The flower is of versicolor")
    elif(5.2<PL<6.2 and 6.2< PW <1.7):
        print("The flower is of verginica")
    else:
        print("The flower is not of Iris type")
    
which_iris()
    
```

```python
counts,bin_edges = np.histogram(df_setosa['petal_length'],bins=10,density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
```

```python
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
